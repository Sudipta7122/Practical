## Practical 4 : Implement Gradient Descent Algorithm to find the local minima of a function.For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
x=2
lr=0.012
precision = 0.000001

previous_step_size= 1
max_iter=10000
iter=0

gf = lambda x:(x+3)**2
```

```python
gd = []
```

```python
while (precision < previous_step_size and iter < max_iter):
    prv = x 
    x = x-lr*gf(prv)
    previous_step_size= abs(x - prv)
    iter+=1
    print("Iteration Value : ", iter, "value :", x)
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
